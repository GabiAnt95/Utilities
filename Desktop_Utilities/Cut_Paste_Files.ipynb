```python
import os
import copy
import sys
import pandas as pd

# Modules for the graphic interface

import tkinter as tk
from tkinter import *
from tkinter import filedialog

from PIL import Image, ImageTk


def resize_image(file, x, y):

    return Image.open(file).resize((x, y))

# -------------------------------------------------- SELECT FILE FUNCTIONS --------------------------------------------------------------------

def files_explorer(variable = 'variable', instructions = "Select a file", multiple = False):
    
    root = tk.Tk()
    
    if multiple == False: 
        file = tk.filedialog.askopenfilename(initialdir= os.getcwd(), title= instructions)
        globals()[variable] = copy.deepcopy(file)
    else: 
        file = tk.filedialog.askopenfilenames(initialdir= os.getcwd(), title= instructions)
        globals()[variable] = [e for e in file]
    
    root.destroy()
    
    return file


def select_new_directory(variable_new_dir = 'variable', instructions = "Select a Directory: "):
    
    root = tk.Tk()
    directory = tk.filedialog.askdirectory(initialdir= os.getcwd(), title= instructions)
    globals()[variable_new_dir] = copy.deepcopy(directory)
    root.destroy()
    return directory

def move(old_directory, new_directory):
    
    os.rename(old_directory, new_directory)
    

def main_move_files():
    
    files = files_explorer(instructions= "Select the Files you want to Move: ", multiple = True)
    new_directory = select_new_directory(instructions= "Select the Destiny Directory: ")

    for file in files:
        old_directory = copy.deepcopy(file)
        only_file = file[len(file) - file[::-1].find('/'):]
        move(old_directory, new_directory + '/' + only_file)
        
def exit_mac():
    window.protocol("WM_DELETE_WINDOW")

    
    
```

```python
window = tk.Tk()
window.title('SELECT EACH INPUT FILE: ')
#window.iconbitmap('')
window.geometry("700x500")
window.config(bg = '#9CDF75')
window.resizable(1,1)

# -------------------------- FRAME ----------------------------------

frame = tk.Frame(window)
frame.config(width = 300, height = 500, bg = 'grey')
frame.pack(expand = 1)
frame.config(bd = '20')


# ------------------------- BUTTONS ----------------------------

#icon_button = ImageTk.PhotoImage(resize_image('images/eye.png', 50, 50))
#button.config(bg = 'green', fg = 'red', font = ("Arial", 12))

# ICONS
icon_exit = ImageTk.PhotoImage(resize_image('images/exit.png', 50, 50))


# BUTTONS
button_projects = Button(frame, text="Select PROJECTS", command=lambda: files_explorer('projects'))
button_pmo_dash = Button(frame, text="Select PMO DASHBOARD", command=lambda: files_explorer('pmo_dash'))
button_finnancials = Button(frame, text="Select FINNANCIALS", command=lambda: files_explorer('finnancials'))
button_tasks = Button(frame, text="Select TASKS", command=lambda: files_explorer('tasks', multiple = True))
button_highligths = Button(frame, text="Select HIGHLIGTHS", command=lambda: files_explorer('highligths'))
button_risks = Button(frame, text="Select RISKS", command=lambda: files_explorer('risks'))

button_output = Button(frame, text = "Select OUTPUT DIRECTORY", command = lambda: select_new_directory('new_directory_output', instructions = "Select OUTPUT DIRECTORY"))
button_exit = Button(frame, text = 'Exit', command = window.destroy, image = icon_exit)

# CONFIGS
button_output.config(bg = 'green')


#GRIDS
button_projects.grid(row = 1, column = 0)
button_pmo_dash.grid(row = 2, column = 0)
button_finnancials.grid(row = 3, column = 0)
button_tasks.grid(row = 4, column = 0)
button_highligths.grid(row = 5, column = 0)
button_risks.grid(row = 6, column = 0)

button_output.grid(row = 7, column = 1)
button_exit.grid(row = 8, column = 1)

window.mainloop()

```
